```python
import numpy as np
import matplotlib.pyplot as plt

#1-a
theta_values = [0, 1, 2, 3, 4, 5]
x_values = np.linspace(0, 1, 1000)

def cdf(x, theta):
    return 1-(( 1-x )**( theta + 1))

for theta in theta_values:
    y_values = cdf(x_values,theta)
    plt.plot(x_values, y_values, label=f"theta = {theta}")

plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Plot of 1-(1-x)^(θ+1) for θ = 0, 1, 2, 3, 4, 5")
plt.legend()
plt.show()

#1-b
print("P(X<0.25 | theta = 2) is :",cdf(0.25,2))

#1-c
np.random.seed(80)
def cdf_inv(x, theta):
    return 1-(( 1-x )**( 1/(theta + 1)))
data = cdf_inv(np.random.uniform(size = 100000),2)
mean = np.mean(data)
variance = np.var(data)

print("Mean of this function is : ", mean)
print("Variance of this function is :", variance)
```
