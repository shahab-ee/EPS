at first we have to import *math,numpy,mathplotlib* library

then we define the *binomial* function that has 3 arguments : n,p,k

we define *bin_values* and* bin_cdf_values* list which they are 0 with size 9

we have a *cdf_value* variable which is cummulated with increasing *k* value from k_values(from 0 to 8 list)

in a loop we initalize the *pmf* to the *bin_values* list and *cdf_value* to the *bin_cdf_values*

then we plot the PMF function with *stem* operation and also CDF function with *step* order

**b:**

we define a variable pk that shows the Pr{X < k } and a counter : k = 0.

in a while loop we increase the pk value by increasing k level by level till pk its < 0.44 and when the loop is finished we have the k which is *Pr{X < k }is > 0.44*

the point is the k which we are looking for is k-1 cause k is the next k which is ok with our condition and its not the smallest

**c:**

in this part we only have to find the sum of the probabilities of k = 3 to 20

so in a for loop we add the pmf values to p3(variable that shows the answer) and then print it.

```python
import numpy as np
import math
import matplotlib.pyplot as plt

#3-a
def binomial(n,p,k):
    return math.comb(n,k)*(p**k)*((1-p)**(n-k))
bin_values = [0]*9
bin_cdf_values = [0]*9
cdf_value = 0
k_values = [0,1,2,3,4,5,6,7,8]
for k in k_values:
   cdf_value += binomial(8,0.3,k)
   bin_values[k] = binomial(8,0.3,k)
   bin_cdf_values[k] = cdf_value


plt.stem(k_values, bin_values)
plt.xlabel("k")
plt.ylabel("P (n,k)")
plt.show()

plt.step(k_values, bin_cdf_values)
plt.xlabel("k")
plt.ylabel("CDF (n,k)")
plt.show()

#3-b
pk = 0
k = 0
while (pk < 0.44):
   pk += binomial(8,0.7,k)
   k +=1

k -= k
print("P (X <= ",k,"is >= 0.44")

#3-c
p3 = 0
for k in range(3,21,1):
    p3 += binomial (20,0.2,k)

print("P ( Y >= 3 ) is : ",p3)
```
